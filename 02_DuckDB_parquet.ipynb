```python
import duckdb

conn = duckdb.connect()

query = """
    SELECT *
    FROM 'data/yellow_trip_data/*.parquet'
    --LIMIT 10
"""

df = conn.sql(query).df()
df
```

```python
# Prima crea una tabella permanente dal DataFrame
conn.execute("CREATE TABLE IF NOT EXISTS my_table AS SELECT * FROM df")

# Poi crea la vista basata sulla tabella
conn.execute("CREATE VIEW IF NOT EXISTS df_view AS SELECT * FROM my_table")
```

```python
import duckdb

query2 = """SELECT * 
            FROM df_view 
            
        """
df = conn.execute(query2).df()
df
```
